```python
# Function to download video or audio
def download_to_google_drive():
    # Ask for the YouTube video URL
    url = input("Please enter the YouTube video URL: ")

    # Ask the user for their choice: 1 for video (MP4), 2 for MP3
    print("Enter your choice:")
    print("1 - Download Video (MP4)")
    print("2 - Download MP3")

    choice = input("Your choice: ").strip()

    # Define Google Drive save path
    drive_path = "/content/drive/My Drive/Youtube Downloads/"

    # Options for downloading as MP4
    video_options = {
        "format": "bestvideo[height<=1080]+bestaudio/best",  # Best quality up to 1080p
        "outtmpl": f"{drive_path}%(title)s.mp4",  # Ensure MP4 extension
        "merge_output_format": "mp4",  # Ensure final format is MP4
        "postprocessors": [
            {
                "key": "FFmpegVideoConvertor",
                "preferedformat": "mp4",  # Convert to MP4
            }
        ],
        "cookiefile": "youtube_cookies.txt",  # Use saved login session
        "progress_hooks": [
            lambda d: print(f"Download Status: {d['status']} - {d.get('filename', '')}")
        ],
    }

    # Options for downloading as MP3
    audio_options = {
        "format": "bestaudio/best",
        "outtmpl": f"{drive_path}%(title)s.mp3",
        "postprocessors": [
            {
                "key": "FFmpegExtractAudio",
                "preferredcodec": "mp3",
                "preferredquality": "192",
            }
        ],
        "cookiefile": "youtube_cookies.txt",
        "progress_hooks": [
            lambda d: print(f"Download Status: {d['status']} - {d.get('filename', '')}")
        ],
    }

    # Select options based on user choice
    if choice == "1":
        options = video_options
        download_type = "MP4 Video"
    elif choice == "2":
        options = audio_options
        download_type = "MP3 Audio"
    else:
        print("Invalid choice. Please enter '1' or '2'.")
        return

    # Download using yt-dlp
    with YoutubeDL(options) as ydl:
        try:
            info = ydl.extract_info(url, download=True)
            print(f"{download_type} Download Complete: {info['title']}")

            # Display file path in Google Drive
            downloaded_file = f"{drive_path}{info['title']}.{'mp4' if choice == '1' else 'mp3'}"
            print(f"File saved in Google Drive at: {downloaded_file}")
        except Exception as e:
            print(f"An error occurred: {e}")

# Connect to Google Drive
connect_to_google_drive()

# Start the download process
download_to_google_drive()
```
